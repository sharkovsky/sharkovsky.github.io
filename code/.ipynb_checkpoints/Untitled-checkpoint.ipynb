```python
n_input = 2

X, y = make_classification(n_samples=800, n_features=n_input, 
                           n_informative=n_input-1, n_redundant=1,
                           n_classes=2, n_clusters_per_class=1, 
                           random_state=42023)
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42 )
```

```python
plt.figure(figsize=(12,12))
plt.scatter(X_train[:,0],X_train[:,1],c=y_train, s=75, linewidths=0)
```

```python
n_hidden_neurons = [8,5]
n_hidden_layers = len(n_hidden_neurons)
nlayers = n_hidden_layers+2
```

```python
def sigmoid(z):
    return 1.0/(1.0 + np.exp(-z))
```

```python
class InputLayer():
    def __init__(self, n=2):
        self.z = np.zeros(shape=(n,1))
    def forward(self, x):
        return x
```

```python
class OutputLayerCrossEntropy():
    def __init__(self, n=1):
        self.N = n
        self.z = np.zeros(shape=(n,1))
        self.b = np.zeros(shape=(n,1))
        self.bupdates = np.zeros(shape=(n,1))
    def forward(self,W,x):
        self.z = np.dot(W,x).reshape(self.N,1) + self.b
        return sigmoid(self.z)
    def last_activ(self):
        return sigmoid(self.z)
    def backward(self, y):
        return ( sigmoid(self.z) - y )*sigmoid(self.z)*( 1.0 - sigmoid(self.z))
    def update_b(self):
        self.b -= self.bupdates
```

```python
class SigmoidLayer():
    def __init__(self, n=5):
        self.N = n
        self.z = np.zeros(shape=(n,1))
        self.b = np.zeros(shape=(n,1))
        self.bupdates = np.zeros(shape=(n,1))
    def forward(self, W, x):
        self.z = np.dot(W,x).reshape(self.N,1) + self.b
        return sigmoid(self.z)
    def last_activ(self):
        return sigmoid(self.z)
    def backward(self, W, dLdz):
        return np.dot(np.transpose(W),dLdz)*sigmoid(self.z)*(1. - sigmoid(self.z))
    def update_b(self):
        self.b -= self.bupdates
```

```python
def cross_entropy(y,a):
    return -y*np.log(a) - (1.-y)*np.log(1. - a)
```

### Weight matrices

```python
Ws = []
Ws.append( np.random.uniform(low=0., high=1., size=(n_hidden_neurons[0],n_input)) )
for i in range(n_hidden_layers-1):
    Ws.append(np.random.uniform(low=0., 
                                high=1., 
                                size=(n_hidden_neurons[i+1],n_hidden_neurons[i])))
Ws.append( np.random.uniform(low=0., high=1., size=(1,n_hidden_neurons[-1])) )
```

```python
layers = []
layers.append( InputLayer(n_input) )
for n in n_hidden_neurons:
    layers.append(SigmoidLayer(n))
layers.append(OutputLayerCrossEntropy())
```

```python
print nlayers
print len(layers)
print len(Ws)
for layer in layers[1:]:
    print layer.z.shape
    print layer.b.shape
```

```python
# Ws[i] is between layers[i] and layers[i+1]
alpha = 0.1
n_epochs = 1000

updates = []
for W in Ws:
    updates.append(np.zeros_like(W))

training_error = []
validation_error = []
    
for i in range(n_epochs):

    errors = []
    for idx, data_sample in enumerate(X_train):
        # feedforward
        a = layers[0].forward(data_sample)
#        print a
        for i in range(1,nlayers):
            a = layers[i].forward(Ws[i-1],a)
#            print a

        errors.append(cross_entropy(y_train[idx], a[0]))
    
        # backprop
        error = layers[nlayers-1].backward(y_train[idx])
        updates[nlayers-2] = np.outer(error, layers[nlayers-2].last_activ() )
        layers[nlayers-1].bupdates = error
       
        for i in range(nlayers-2,1,-1):
            error = layers[i].backward(Ws[i], error)
            updates[i-1] = np.outer(error , layers[i-1].last_activ())
            layers[i].bupdates = error

        for i,update in enumerate(updates):
            Ws[i] -= alpha*update
            
        for i in range(1,nlayers):
            layers[i].update_b()
            
    training_error.append(np.mean(errors))
    
    # validation
    errors = []
    for idx, data_sample in enumerate(X_test):
        # feedforward
        a = layers[0].forward(data_sample)
        for i in range(1,nlayers):
            a = layers[i].forward(Ws[i-1],a)
        errors.append(cross_entropy(y_test[idx], a[0]))
    validation_error.append(np.mean(errors))
            
#        assert False
```

```python
plt.figure(figsize=(12,12))
plt.plot(range(n_epochs), training_error, '.-')
plt.plot(range(n_epochs), validation_error, '.-')
plt.legend(['training','validation'])
```

```python
def inference(layers, Ws, x):
    # feedforward
    a = layers[0].forward(x)
    for i in range(1,nlayers):
        a = layers[i].forward(Ws[i-1],a)
        
    return a > 0.5
```
